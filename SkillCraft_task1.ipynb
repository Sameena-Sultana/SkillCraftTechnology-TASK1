```python
# BAR CHART : AVERAGE TEMPERATURE BY SEASON 
plt.figure(figsize=(6,4))
sns.barplot(
    x="season",
    y="temperature",
    data=df,
    order=["Spring","Summer","Monsoon","Autumn","Winter"],
    estimator="mean",
    errorbar=None
)
plt.ylabel("Mean Temperature (°C)")
plt.title("Average Temperature by Season – Hyderabad")
plt.tight_layout()
plt.show()
```

```python
# BAR CHART: AVERAGE POLLUTION LEVELS BY POLLUTION SOURCE
plt.figure(figsize=(6,4))
sns.barplot(x="pollution_source",
            y="pollution_level",
            data=df,
            estimator="mean",
            errorbar=None,
            color="#8e44ad")  # Plum Purple
plt.ylabel("Mean Pollution Level (μg/m³)")
plt.title("Average Pollution Levels by Source – Hyderabad")
plt.show()
```

```python
# MONTHLY TEMPERATURE HISTOGRAM
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="temperature",
    hue="month",
    element="step",
    stat="count",
    common_norm=False,
    bins=20
)
plt.xlabel("Temperature (°C)")
plt.title("Monthly Temperature Histogram – Hyderabad")
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM:  RAINFALL FREQUENCY DISTRIBUTION (STACKED) ACROSS 12 MONTHS
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="rainfall",
    hue="month",
    multiple="stack",
    bins=15,
    edgecolor=".3"
)
plt.xlabel("Rainfall (mm)")
plt.title("Rainfall Frequency Distribution Across Months – Hyderabad")
plt.tight_layout()
plt.show()
```
